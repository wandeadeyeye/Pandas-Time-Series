### Pandas Time Series Tutorial: DateTimeIndex

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv('https://query1.finance.yahoo.com/v7/finance/download/AAPL?period1=1593937224&period2=1625473224&interval=1d&events=history&includeAdjustedClose=true',sep=',',parse_dates=['Date'],index_col='Date')
```

```python
data.index
```

```python
data.Close.resample('M').mean() #monthly average of closed price
```

```python
data.Close.resample('M').mean().plot()
plt.title('Monthly Average of Closed Price')
```

```python
data.Close.resample('W').mean().plot()
plt.title('Weekly Average of Closed Price')
```

```python
data.Close.resample('Q').mean().plot() #quater end frequency
plt.title('Quater End Frequency')
```

```python
data.Close.resample('M').mean().plot(kind = 'bar')
plt.title('Monthly Average Frequency')
```

```python
data.Close.plot()
plt.title('Closing Price Chart')
```
