## Pandas Time Series Analysis Tutorial: shifting & lagging

```python
import pandas as pd
```

```python
data = pd.read_csv('C:/github/fb.csv',parse_dates=['Date'],index_col='Date')
data
```

### Shift

```python
data.shift(1)
```

```python
data.shift(-1)
```

```python
data['Prev Day Price'] = data['Price'].shift(1)
data
```

```python
data['Price Change'] = data['Price'] - data['Prev Day Price']
data
```

```python
data['5 day return'] =  (data['Price'] - data['Price'].shift(5))*100/data['Price'].shift(5)
data
```

```python
data = data[['Price']]
data
```
