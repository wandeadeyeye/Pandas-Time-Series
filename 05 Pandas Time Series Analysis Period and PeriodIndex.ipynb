# Pandas Time Series Analysis: Period and PeriodIndex

```python
import pandas as pd
import numpy as np
```

### Yearly Period

```python
#yearly time period
y = pd.Period('2020')
y
```

```python
dir(y)
```

```python
y.start_time
```

```python
y.end_time
```

## Monthly Period

```python
#monthly time period
m = pd.Period('2019-1',freq='M')
m
```

```python
m.start_time
```

```python
m.end_time
```

```python
m + 1
```

## Daily Period

```python
# daily time period
d = pd.Period('2020-09-25',freq='D')
d
```

```python
d + 1
```

## Hourly Period

```python
# hourly frequency
h = pd.Period('2020-09-25 22:00:00',freq='H')
h
```

```python
h.start_time
```

```python
h.end_time
```

```python
h + 1
```

## Achieve same results using pandas offsets hour

```python
h+pd.offsets.Hour(2)
```

## Quarterly Period

```python
# quaterly frequency
q = pd.Period('2020Q1')
q
```

```python
q+1
```

```python
# quaterly frequency for fiscal year
q = pd.Period('2020Q1',freq='Q-JAN')
q
```

```python
q.start_time
```

```python
q.end_time
```

## Use asfreq to convert period to a different frequency

```python
q.asfreq('M',how='start')
```

```python
q.asfreq('M',how='end')
```

```python
q2 = pd.Period('2021Q2',freq='Q-JAN')
q2
```

```python
q2-q
```

## Weekly Period

```python
w = pd.Period('2019-07-05',freq='W')
w
```

```python
w-1
```

```python
w2 = pd.Period('2019-08-15',freq='W')
w2
```

```python
w2-w
```

### PeriodIndex and period_range

```python
# creating a period index
idx = pd.period_range('2011','2017',freq='Q')
idx
```

```python
idx[0].start_time
```

```python
idx[0].end_time
```

#### Walmart's fiscal year ends in Jan, below is how you generate walmart's fiscal quarters between 2011 and 2017

```python
# creating a fiscal period index
idxF = pd.period_range('2011','2017',freq='Q-JAN')
idxF
```

```python
idxF[0].start_time
```

```python
idx[0].end_time
```
