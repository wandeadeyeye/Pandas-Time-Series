### Pandas Time Series Analysis Tutorial: date_range

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
data = pd.read_csv('https://query1.finance.yahoo.com/v7/finance/download/AAPL?period1=1593937224&period2=1625473224&interval=1d&events=history&includeAdjustedClose=true',sep=',')
```

```python
data.head()
```

```python
data.drop(columns=['Date'], inplace=True)
data.head()
```

```python
data.shape
```

```python
rang = pd.date_range(start='07/06/2020',periods=252,freq='B')
rang
```

```python
data.set_index(rang,inplace=True)
data.head()
```

### Finding missing dates from datetimeindex

```python
daily_index = pd.date_range(start="7/6/2020",periods=252,freq='D')
daily_index
```

```python
daily_index.difference(data.index)
```

```python
data['2020-07-06':'2020-08-07']
```

### Benefits of having DatetimeIndex

```python
data['2020-08-01':'2020-08-31'].Close.plot()
plt.title('Close')
```

```python
data['2020-08-01':'2020-08-31'].Close.mean()
```
